```python
import matplotlib.pyplot as plt
import time
from IPython import display
import math
import random
import pandas as pd
```

```python
X1=[0,0,1,1]
X2=[0,1,0,1]
Y=[0,0,0,1] # AND Gate
```

```python
plt.figure(figsize=(4,2))
for i in range(len(Y)):
  if Y[i]==0 :
    plt.plot(X1[i],X2[i],'ro')
  else:
    plt.plot(X1[i],X2[i],'go')
plt.show()
```

```python
W=[random.random(),random.random()]
b=random.random()
print(W,b)

lr=0.01
threshold=0.5
```

```python
epoch = 0
count = 0
loss=[]
while count < 4:
  total_error=0
  for i in range(len(Y)):
    pred=0
    weighted_sum = W[0]*X1[i] + W[1]*X2[i] + b
    if weighted_sum > threshold:
      pred=1
    error= Y[i] - pred
    if error!=0:
      W[0] += lr*error*X1[i]
      W[1] += lr*error*X2[i]
      b+=lr*error
      count=0
      total_error+=error
    else:
      count+=1
  loss.append(0.25*total_error)
  plt.figure(figsize=(8,4))
  display.clear_output(wait=True)
  plt.subplot(1,2,1)
  for i in range(len(Y)):
    if Y[i] == 0:
      plt.plot(X1[i],X2[i],'ro')
    else:
      plt.plot(X1[i],X2[i],'go')
  plt.xlim([-1,2])
  plt.ylim([-1,2])

  xintr=(threshold-b)/W[0]
  yintr=(threshold-b)/W[1]
  slope=-yintr/xintr
  xx=plt.xlim()
  yeq=[]
  for x in xx:
    yeq.append(slope*x+yintr)
  plt.plot(xx,yeq)

  plt.subplot(1,2,2)
  plt.plot(loss)
  plt.tight_layout(pad=3)
  plt.show()
  time.sleep(0.2)
```

```python
W = [random.random(), random.random()]
b = random.random()
lr = 0.1
threshold=0.5
loss=[]
count=0
epoch=0
while count<99:
  total_error=0
  for i in range(len(X1)):
    weighted_sum=W[0]*X1[i]+W[1]*X2[i]+b
    pred=0
    if weighted_sum > threshold :
      pred=1
    error = Y[i] - pred
    if error !=0:
      count=0
      total_error+=error
      W[0]+=lr*error*X1[i]
      W[1]+=lr*error*X2[i]
      b+=lr*error
    else:
      count+=1
  loss.append(0.01*total_error)
  plt.figure(figsize=(10,4))
  display.clear_output(wait=True)
  plt.subplot(1,2,1)
  plt.title(f"Epoch: {epoch}")
  for i in range(len(Y)):
    if Y[i] == 0:
      plt.plot(X1[i],X2[i],'ro')
    else:
      plt.plot(X1[i],X2[i],'go')
  plt.xlim([-10,120])
  plt.ylim([-10,120])

  xintr=(threshold-b)/W[0]
  yintr=(threshold-b)/W[1]
  slope=-yintr/xintr
  xx=plt.xlim()
  yeq=[]
  for x in xx:
    yeq.append(slope*x+yintr)
  plt.plot(xx,yeq)

  plt.subplot(1,2,2)
  plt.plot(loss)
  plt.tight_layout(pad=3)
  plt.show()
  time.sleep(0.01)
  epoch+=1

```
